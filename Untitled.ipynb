```python
import sqlite3
import pandas as pd

sql = sqlite3.connect('system_of_orders.db')
```

```python
pd.read_sql(
    """
    CREATE TABLE emp (
  idemp int(11) NOT NULL,
  fam varchar(60) NOT NULL,
  name varchar(60) NOT NULL,
  fname varchar(60) NOT NULL,
  birthdate date NOT NULL,
  email varchar(45) DEFAULT NULL,
  PRIMARY KEY (idemp)
)
    """,
    sql)
```

```python
pd.read_sql(
    """
    CREATE TABLE dolz (
  iddolz int(11) NOT NULL,
  dolname varchar(100) NOT NULL,
  PRIMARY KEY (iddolz)
)
    """,
    sql)
```

```python
pd.read_sql(
    """
    CREATE TABLE otd (
  idotd int(11) NOT NULL,
  sokr_naz varchar(100) NOT NULL,
  naz varchar(100) NOT NULL,
  par_otd int(11) NOT NULL,
  PRIMARY KEY (idotd),
  CONSTRAINT par_otd FOREIGN KEY (par_otd) REFERENCES otd (idotd) ON DELETE NO ACTION ON UPDATE NO ACTION
)
    """,
    sql)
```

```python
pd.read_sql(
    """
    CREATE TABLE dol_emp (
  id int(11) NOT NULL,
  id_emp int(11) NOT NULL,
  id_dol int(11) NOT NULL,
  id_podr int(11) NOT NULL,
  type_dol float DEFAULT NULL,
  stavka float DEFAULT NULL,
  date_priem date DEFAULT NULL,
  date_uvol date DEFAULT NULL,
  status int(11) NOT NULL,
  PRIMARY KEY (id_emp,id_dol,id_podr,id),
  CONSTRAINT id_dol FOREIGN KEY (id_dol) REFERENCES dolz (iddolz) ON DELETE CASCADE ON UPDATE CASCADE,
  CONSTRAINT id_emp FOREIGN KEY (id_emp) REFERENCES emp (idemp) ON DELETE CASCADE ON UPDATE CASCADE,
  CONSTRAINT id_podr FOREIGN KEY (id_podr) REFERENCES otd (idotd) ON DELETE CASCADE ON UPDATE CASCADE
)
    """,
    sql)
```
